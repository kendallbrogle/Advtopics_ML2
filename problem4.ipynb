# Problem 4 - Linear Separability

## 4.1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
# Define the dataset points
class1 = np.array([[-1, -1], [1, 1], [-3, -3], [4, 4]])
class2 = np.array([[-1, 1], [1, -1], [-5, 2], [4, -8]])

#plot the pts
plt.scatter(class1[:, 0], class1[:, 1], label='Class 1', color = "green", marker='o')
plt.scatter(class2[:, 0], class2[:, 1], label='Class 2', marker='x')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Dataset Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Original dataset
X = np.array([[-1, -1], [1, 1], [-3, -3], [4, 4], [-1, 1], [1, -1], [-5, 2], [4, -8]])
y = np.array([1, 1, 1, 1, 0, 0, 0, 0])

# Train a linear SVM
clf = SVC(kernel='linear')
clf.fit(X, y)

# Check if it classifies correctly
right_classif = np.all(clf.predict(X) == y)
print(right_classif)
```

**Answer:**

This is not linearly seperable seen in the plot and using an SVM classifier.

## 4.2

```python
#  new 1-dimensional representation z = x1 * x2 for both classes
z1 = np.prod(class1, axis=1)
z2 = np.prod(class2, axis=1)

# Plot the dataset in the z-space
plt.scatter(z1, np.zeros_like(z1), label='Class 1', color='green', marker='o')
plt.scatter(z2, np.zeros_like(z2), label='Class 2', marker='x')
plt.xlabel('z = x1 * x2')
plt.title('Dataset Plot in z-Space')
plt.legend()
plt.grid(True)
plt.show()
```

**Answer:**
To make the dataset linearly separable, we can define a new 1-dimensional representation, z, that transforms the data in such a way that it becomes linearly separable. The transformation z is defined as the product of x1 and x2 for both classes.

By applying this transformation, we create new representations for each class, z1 and z2, which are defined as the products of x1 and x2 for the respective classes.

In this 1-dimensional representation space, we can draw a single straight line that separates the two classes cleanly. Class 1 is clustered around one side of the line, and Class 2 is clustered around the other side.


## 4.3

```python
#  new 1-dimensional representation z = x1 * x2 for both classes
z1 = np.prod(class1, axis=1)
z2 = np.prod(class2, axis=1)

# Plot the dataset in the z-space
plt.scatter(z1, np.zeros_like(z1), label='Class 1', color='green', marker='o')
plt.scatter(z2, np.zeros_like(z2), label='Class 2', marker='x')
plt.axvline(x=0, color='red', linestyle='--', label='Separating Hyperplane (z = 0)')
plt.xlabel('z = x1 * x2')
plt.title('Dataset Plot in z-Space')
plt.legend()
plt.grid(True)
plt.show()
```

**Answer:**
Now linearly seperable where line Z = 0 based on a x1x2 transformation.
